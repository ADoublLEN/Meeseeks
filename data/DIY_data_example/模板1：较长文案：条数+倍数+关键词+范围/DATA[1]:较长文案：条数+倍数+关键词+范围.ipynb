```python
# data 1.1
import json, ast, random, re

def txt_to_json(text):
    return ast.literal_eval(text)

with open("/Users/jiamingwang/Documents/指令遵循/tuanbench2/tuanbanch/ComplexBench-qwen/新版合成数据模板/模板/模板1：较长文案：条数+倍数+关键词+范围/data1.1.json", "r") as f:
    data = json.load(f)

def calculate_and_replace(text):
    # 找到匹配 ###A*B*C### 的表达式
    pattern = r'###(.*?)###'
    
    def evaluate_expression(match):
        expression = match.group(1)
        try:
            # 使用eval计算表达式的值
            result = eval(expression)
            # 保留最多一位小数
            formatted_result = "{:.1f}".format(result)
            return formatted_result
        except:
            return match.group(0)
    
    # 用计算后的结果替换原始表达式
    new_text = re.sub(pattern, evaluate_expression, text)
    return new_text

item_nums = [4, 7]
mults = [0.5, 1.2]
keywords = ["富士山", "自然美景", "日本文化", "冒险", "温泉", "美食", "摄影", "旅行体验", "景点推荐"]
non_keywords = ["适合", "快乐", "轻松", "旅游", "国外", "的", "了", "你", "我"]


item1 = data[0]
keys = []
for _ in range(2):
    keyw_sublist = random.sample(keywords, 8)
    non_keyw_sublist = random.sample(non_keywords, 8)
    keyw_res = ""
    non_keyw_res = ""
    for item in keyw_sublist:
        keyw_res = ", ".join(f"\"{item}\"" for item in keyw_sublist)
    for item in non_keyw_sublist:
        non_keyw_res = ", ".join(f"\"{item}\"" for item in non_keyw_sublist)
    keys.append([keyw_res, non_keyw_res])

new_data = []
def func(item):
    new_data = []
    for item_num in item_nums:
        for keyword in keys:
            for mult in mults:
                # 在每次的最内层循环中生成一个新的项
                new_item = str(item)
                new_item = new_item.replace("###条数###", str(item_num))
                new_item = new_item.replace("###关键词###", str(keyword[0]))
                new_item = new_item.replace("###反选关键词###", str(keyword[1]))
                new_item = new_item.replace("###倍数###", str(mult))
                new_item = new_item.replace("【倍数】", str(mult))
                new_item = calculate_and_replace(new_item)
                new_data.append(txt_to_json(new_item))

    return new_data 

new_data += func(item1)
# new_data += func(item2, result1)

for item in new_data:
    for subq in item["sub_questions"]:
        if "能力项" in subq:
            text = subq["能力项"]
            capatags = text.split("、")
            for capatag in capatags:
                if capatag[-1] == "字":
                    num = float(capatag[:-1])
                    if 0 <= num < 10:
                        replacement = "0~10字"
                    elif 10 <= num < 50:
                        replacement = "10~50字"
                    elif 50 <= num < 200:
                        replacement = "50~200字"
                    elif num >= 200:
                        replacement = "200字以上"
                    subq["能力项"] = text.replace(capatag, replacement)
            

print(len(new_data))
with open("/Users/jiamingwang/Documents/指令遵循/tuanbench2/tuanbanch/ComplexBench-qwen/新版合成数据模板/数据/DATA[1.1].json", "w") as f:
    json.dump(new_data, f, indent=4, ensure_ascii=False)
```
